Closing Statement (three critical caveats)
First, the flowed-through data barely varies the disputed parameter: d/D only spans 0.017–0.039 and ψ only 0.38–0.42 (§5.5). The d/D controversy cannot be settled from Zehner's flowed data  alone; it would be necessary to fold in the other-author datasets the thesis already cites (Fahien–Smith, Seidel, etc., Abb. 62) or generate new points.
Second, the data are a single-source 1972 scan; digitization error and apparatus-specific bias are part of the uncertainty budget.
Third, replacing the VDI free/mixed-convection correlation patchwork with one neural model for pure accuracy gains by enforcing smoothness and reporting UQ is unimpressive: the Wärmeatlas admits the discontinuities at regime boundaries (Fc: "Bei Winkeln α = 0° und 90° … Unsicherheit der angegebenen Beziehungen").
Likewise, a DeepONet/FNO surrogate of the already-analytic λso map, or a large deep net on this small tabular set, is the wrong tool: with only a few hundred single-apparatus rows and ~6–8 physical features, deep learning will overfit and forfeit interpretability.
